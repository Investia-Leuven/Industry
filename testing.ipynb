```python
from yahooquery import Screener
import pandas as pd

s = Screener()

# Try 'basic_materials' or specific industries like 'gold'
screen_id = 'consumer_electronics'  # or 'agricultural_chemicals', etc.
results = s.get_screeners(screen_id, count=500)

# Structure: results[screen_id]['quotes']
if screen_id in results and 'quotes' in results[screen_id]:
    comps = results[screen_id]['quotes']
else:
    print("No data — check available_screeners for valid screener ID")
    comps = []

df = pd.DataFrame(comps)
print(df.head())
```
